6. Al hacer una tabla de frecuencia sobre la cantidad total de camas críticas en los hospitales de la red, considerando intervalos de 3 en 3, ¿en cuántos intervalos hay exactamente 1 cama crítica?  

```python
df_hospitales[["Nombre Establecimiento","Total Camas Críticas"]].plot(kind="hist", bins=3)
```

```python


# Suponiendo que 'df_hospitales' es tu DataFrame y que la columna de interés se llama "Total Camas Críticas"

# 1. Definir el rango de los bins (intervalos) de 3 en 3.
min_val = df_hospitales["Total Camas Críticas"].min()
max_val = df_hospitales["Total Camas Críticas"].max()

# Asegurarse de que el máximo se incluya; por eso se suma 3
bins = np.arange(min_val, max_val + 3, 3)

# 2. Crear la tabla de frecuencia usando pd.cut para agrupar en los intervalos
freq_table = pd.cut(df_hospitales["Total Camas Críticas"], bins=bins, right=False).value_counts().sort_index()

# Mostrar la tabla de frecuencia
print("Tabla de frecuencia:")
print(freq_table)

# 3. Contar cuántos intervalos tienen exactamente 1 hospital (o 1 registro)
intervals_with_one = (freq_table == 1).sum()

print("\nNúmero de intervalos con exactamente 1 cama crítica:", intervals_with_one)
```
